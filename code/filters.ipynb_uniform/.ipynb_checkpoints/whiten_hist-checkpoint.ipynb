```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color, exposure

%matplotlib inline
def whiten(img):
    mu = np.mean(img)
    std = np.std(img)
    return (img - mu) / std
```

```python
img = Image.open('/Users/kuoyen/Desktop/beachsamp1.jpg')
plt.imshow(img)
```

```python
img = whiten(img)
```

```python
plt.imshow(img)
```

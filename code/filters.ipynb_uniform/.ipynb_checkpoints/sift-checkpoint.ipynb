```python
import cv2
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
%matplotlib inline
from skimage.color import rgb2gray

img = io.imread('/Users/kuoyen/Desktop/beachsamp1.jpg')
# gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gray = rgb2gray(img)

sift = cv2.SIFT()
kp = sift.detect(gray,None)

img=cv2.drawKeypoints(gray,kp)

cv2.imwrite('/Users/kuoyen/Desktop/sift_keypoints.jpg',img)
```
